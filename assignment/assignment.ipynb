1. This paper discusses the idea of how data cleaning process can be optimized to be easier and effective through data tidying. This idea of tidying makes it easier to manipulate and visualize datasets by having a structure of variables being columns, observations as rows, and observational units as tables.

2. It intends to establish a framework with how to tidy messy datasets of all types without having to utilize too many tools and creating a new process to clean new data every time. This allows for users to focus more on the exploration and analysis of the data.

3. In the first sentence, the phrase "Like families, tidy datasets are alike" means that tidy datasets tend to follow the same principles such as the structure mentioned in the abstract. The rest of that sentence explains that though they follow that structure, each dataset may have its own problems such as duplicate values, missing values or other errors. The second sentence describes how based on a given dataset, using the structure given in the abstract, it's easy to understand what each piece of data represents. However, it is very difficult to provide a universal definition of variables and observations as an observation in one dataset can be considered a variable in another as studies can be taken in different contexts.

4. A value is either a number or a string that belongs to a variable and oand bservation. A variable is a collection of values that measure the same attribute across units. An observation is a collection of values measured across the same unit over multiple attributes.

5. Tidy data is based on 3 characteristics: Each variable forms a column, each observation forms a row, and each observational unit forms a table.

6. The 5 most common problems with messy datasets are: Column headers are values and not variable names, multiple variables are stored in one column, variables are stored in both rows and columns, multiple types of obersvational units are stored in the same table, and a single observational unit is stored in multiple tables. The data in table 4 is considered messy because it variables form both rows and columns, and the columns are values that measure income instead of being variable names such as "Agnostic". "Melting" a dataset means turning columns into rows.

```python
url = 'http://www.vcsc.virginia.gov/pretrialdataproject/October%202017%20Cohort_Virginia%20Pretrial%20Data%20Project_Deidentified%20FINAL%20Update_10272021.csv'
df = pd.read_csv(url,low_memory=False) # Pandas downloads and loads the .csv file for you
df.head()
```

```python
wsdrp = df['WhetherDefendantWasReleasedPretrial']
print(wsdrp.value_counts())
wsdrp = wsdrp.replace([9], np.nan)
df['WhetherDefendantWasReleasedPretrial'] = wsdrp
print(df['WhetherDefendantWasReleasedPretrial'].value_counts)
```

Part 4

```python
isacice = df['ImposedSentenceAllChargeInContactEvent']
stacacice = df['SentenceTypeAllChargesAtConvictionInContactEvent']
print(stacacice.value_counts())
print(isacice.unique())
```

```python
isacice = pd.to_numeric(isacice, errors='coerce') #used this to convert all values into integers
dummy = isacice.isnull() # kept track of null values or missing values
print( pd.crosstab(dummy, stacacice), '\n') #original crosstab shows that there are values missing for 4 and 9
```

```python
isacice = isacice.mask( stacacice == 4, 0) #replace this so that we can manage most of the values
isacice = isacice.mask( stacacice == 9, np.nan)
dummy = isacice.isnull()
print(pd.crosstab(dummy, stacacice), '\n') # new crosstab to verify previous operations
df['ImposedSentenceAllChargeInContactEvent'] = isacice
del isacice, stacacice
```
